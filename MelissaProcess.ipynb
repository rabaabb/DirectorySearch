```python
#this is where it takes in the input - a excel or csv file with 
#Last name , Zip Code (as columns, in that order)


#this takes in an excel file input 
df1 = pd.read_excel(r'/Users/aishvishah/Documents/brightfox/Cal Excel/2025/Cal2025-12.xlsx')

#take a csv file input - change the red part 
#df1 = pd.read_csv('file_name.csv')
df1
```

```python
links = []


#edit this code for different zip codes and cities 
zip_code = "94583"
city = "San+Ramon"

for index in range(len(df1)):
    last_name = df1.iloc[index][0][:-1]
    link = "https://www.melissa.com/v2/lookups/personatorsearch/?name=" + last_name + "&city=" + city + "&state=CA&postalCode=" + zip_code + "&melissaAddressKey=&phoneNumber=+&emailAddress=+&freeForm=&fmt=json&id=NIJVq8ms8hazqSiCXu_srO**nSAcwXpxhQ0PC2lXxuDAZ-**"
    links.append(link)
```

```python
#make this a method that it calls each time, for all valid links 
def create_panda(html):
    #initializing all the lists to hold data
    full_names = []
    first_names = []
    last_names = []
    addresses = []
    ages = []

    pattern = "FullName"
    #run this code to append all FullNames
    for match in re.finditer(pattern, html):
        starting = match.end() + 3
        ending = match.end() + 30
        full_name = html[starting : ending]
        full_name = full_name.split("\",")[0]
        full_names.append(full_name)


    #run this code to append all FirstName
    pattern1 = "FirstName"
    for match in re.finditer(pattern1, html):
        starting = match.end() + 3
        ending = match.end() + 30
        first_name = html[starting : ending]
        first_name = first_name.split("\",")[0]
        first_names.append(first_name)


    #run this code to append all LastName
    pattern2 = "LastName"
    for match in re.finditer(pattern2, html):
        starting = match.end() + 3
        ending = match.end() + 30
        last_name = html[starting : ending]
        last_name = last_name.split("\",")[0]
        last_names.append(last_name)

    #run this code to append all AddressLine1
    pattern3 = "AddressLine1"
    for match in re.finditer(pattern3, html):
        starting = match.end() + 3
        ending = match.end() + 30
        address = html[starting : ending]
        address = address.split("\",")[0]
        addresses.append(address)

    #run this code to append all Age
    pattern4 = "Age"
    for match in re.finditer(pattern4, html):
        starting = match.end() + 3
        ending = match.end() + 5
        age = html[starting : ending]
        age = age.split("\",")[0]
        ages.append(age) 

    tiny_df = pd.DataFrame({"FullName": full_names,
                       "FirstName": first_names, 
                      "LastName": last_names, 
                      "AddressLine1": addresses, 
                      "Age": ages})
    return tiny_df




```

```python
#initializing the data frame that stores all the information 
df_with_info = pd.DataFrame(columns = ['FullName', 'FirstName', 'LastName', 'AddressLine1', 'Age'])
df_with_info
```

```python
#for loop to traverse all the links 
for url in links:
    #url = "https://www.melissa.com/v2/lookups/personatorsearch/?name=Le&city=San+Ramon&state=CA&postalCode=94583&melissaAddressKey=&phoneNumber=+&emailAddress=+&freeForm=&fmt=json&id=NIJVq8ms8hazqSiCXu_srO**nSAcwXpxhQ0PC2lXxuDAZ-**"
    req = requests.get(url)
    if req.status_code in [200]:
        html = req.text
        df_returned = create_panda(html)
        #concatonate all the dataframes to create a big one

        if not df_returned.empty:
            print(df_returned)
            mix = [df_returned, df_with_info]
            df_with_info = pd.concat(mix)
            df_with_info
    else:
        continue
```
